### Abstract

This statistical study has been carried out to find the relation between various price indicators on the closing price of the cryptocurrency Bitcoin. For this, a multiple regression model was used. There is significant evidence that suggests that the opening-price, high-price and low-price are predictors for the closing-price. It is remarkable that volume is not statistically significant to influence the closing-price. Eventually, this model can be used to predict at what price Bitcoin will close on a given day.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from scipy.stats import probplot
import warnings
# We import the warnings library to stop warnings from showing up as they do not provide added insights into statistical analysis
warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
```

### Dataset

We found a dataset that represents historical data of numerous cryptocurrencies ("coin.csv"). For our report, we will only be focusing on predicting the Bitcoin price, as we believe it is the most important coin in the market and potentially even drives the market. For this, we will be using a dataset containing a subset of coin.csv: "coin_Bitcoin.csv". This dataset contains the opening price, closing price, daily high, daily low, daily volume and market cap per given day. Our data spans over a period of approximately 8 years, where each data point represents a specific day within that period.

We retrieved our data set as a subset from the Kaggle dataset "Cryptocurrencies Historical Prices". Kaggle states that all data has been collected from coinmarketcap.com and is therefore free to use. Our dataset is highly rated, receiving a rating of 9.7 for usability. After examining our dataset, we determined we will not need every key available to us. Therefore, we will be dropping two variables from our dataset:

* Firstly, we will drop the "Name" variable, as we will only be using the Bitcoin cryptocurrency, so we will not need any additional names beside "Bitcoin". 
* Secondly, we will the "Symbol" variable, as we will only be using the Bitcoin cryptocurrency, so we will not need any additional coin identifier.
* Finally, we decided to drop rows in which there was no volume measured. We assume that this is a mistake in the data, as other sources do suggest that on these dates some trades were made.

Based on our results, we will attempt to make a general claim about what variables are influential when trying to predict the closing-price of cryptocurrencies. We believe that due to the similarities between cryptocurrencies the underlying fundamental influences are the same. This means that we expect that what influences the price of Bitcoin will likely also be an influence on other cryptocurrencies. Because of this, we can make careful claims about the complete population of cryptocurrencies, which is rumoured to be 10.000+ types of cryptocurrencies.

```
data = pd.read_csv("coin_Bitcoin.csv")

# We remove both the name and the symbol variable, as these are identical across all data points. 
# We will also drop the SNo variable as this keeps track of how many datapoints there are, but this is also done by the rows in excel
# We will drop rows in which a volume of 0 was recorded for reasons explained above
data = data.drop("Name", axis=1)
data = data.drop("Symbol", axis=1)
data = data.drop("SNo", axis=1)
no_volume = data[data["Volume"] == 0].index
data = data.drop(no_volume, axis=0)

print("The dataset contains {} datapoints".format(len(data['Close'])))
for key in data.keys():
    print(" {0} : {1} missing value(s)".format(key, data[key].isna().sum()))   
print("{} rows were dropped because there was a value of 0 for volume".format(len(no_volume)))     
```

### Response variable
The response variable that we will try to predict using multiple-regression is the closing price in Dollars. Below we plotted the values that our response variable can take. It is important to note that the closing-price fluctuates a lot, and due to this it is hard to predict in what range we can realistically expect our response variable to fall. Based on the current consensus among experts, bitcoin is likely to become worth more over time, however it is impossible to accurately predict at what value it will stop. Because of that, we set the range between \$0 and \$65.000. With this it is important to note that \$65.000 dollars is roughly the current ATH, but it is expected to become much more than this. It is important to note that in the plot below, the x-axis is corresponding to the order of collection. The first datapoint was collected on 29/04/2013 and the final one on 27/02/2021.

When looking at the plot of the response variable, there are two noticable peaks. The first around datapoint 1700, the second one seems to be happening currently. Scarcity in economics nearly always leads to an increase in price, and due to this increase people still feel incentivised to mine Bitcoin. While this cycle seems to be a good indicator of changes over a longer period of time, it does not seem to accurately explain changes on a given day. That is why our model is not focused on historical data, but only data of the same day.
```
sns.lineplot(data = data["Close"])
plt.title("Closing price in Dollars against order of collection")
plt.xlabel("Order of collection following date")
plt.ylabel("Closing price in Dollars")
plt.show()
print("Figure 1: A line graph of closing price in Dollars against order of collection following date")
```

### Hypothesis

$H_0$: The closing price cannot be predicted by opening-, high-, low-price, volume and marketcap

$H_A$: The closing price can be predicted by opening-, high-, low-price, volume and marketcap

For every abovementioned dependent variable, where $\beta_i$ represents the coefficient of variable i in the predicted line:

$H_0$: $ \beta_i = 0$ when the other explanatory variables are included in the model

$H_A$: $ \beta_i \neq 0$ when the other explanatory variables are included in the model

### Methods
To access the validity of our hypothesis, we will be using backwards-regression to construct a multiple-regression model. For this, we will be using a significance level of $\alpha = 0.05$, we will be using the adjusted $R^2$ as a tiebreaker in case multiple variables have the same p-value. For our multiple-regression model, we consider the variables high-price, low-price, opening-price, volume and marketcap

```
model = sm.formula.ols(formula= "Close ~ High + Low + Open", data= data)
multi_reg = model.fit()
print(multi_reg.summary())
```

Based on the results of this model, we have a strong suspision of multicollinearity. Because of this, it is very important to thoroughly investigate the relation between our explanatory variables.

Based on the histogram displayed below, the first thing that is very clear, is that there is a massive peak around a value of 0 for the residuals. We also notice that the residuals seem to follow a fairly normal distribution, though it is important to point out that it seems like the left-tail of the distribution drops more gradually to a value of 0 density, while the right-tail seems to drop much more suddenly. When we look at the QQ-plot, the plot indicates that there is a big peak in the middle of the distribution. It is also important to note that at both the beginning and the end of the plot we deviate from the "normal" line. Based on both the QQ-plot, and skinny histogram, we have indication that our kurtosis may be very high. And indeed, when checking this in our model summary above, we find a kurtosis of 75.074.
Because this value is so high, we have more data in our tails than a normal distribution, so we can not confidently claim that our residuals are normally distibuted. Due to this, we must be very careful when discussing our final model and mention its potential limitations. 

```
predicted_value = multi_reg.predict()
residuals = data['Close']-predicted_value

plt.figure(figsize=(8,8))
plt.subplot(121)
probplot(residuals, plot=plt)

plt.subplot(122)
sns.distplot(residuals, bins=100)
plt.title("Histogram of the residuals")
plt.xlabel("Residuals")
plt.show()
print("Figure 2: A probability plot (left) and a histogram (right) of the residuals of the coin_Bitcoin dataset")
```

```
fig, (ax1) = plt.subplots(1, figsize=(10,10))

ax1.scatter(x = predicted_value, y = residuals)
ax1.set_title("True values vs Predicted values")
ax1.set_xlabel("Predicted values")
ax1.set_ylabel("Residuals")
plt.show()
print("Figure 3: A scatterplot of the residuals absolute value against the true value of the data from the coin_Bitcoin dataset")
```

Based on the above scatterplot, we see a big cloud between a predicted value of \$0 and \$20.000 centered around 0. After it has crossed the 20.000 mark, the data seems to be more randomly scattered. Based on this, we assume that the closing price was easier to predict in the beginning of bitcoins lifespan. This also makes sense if we look at the data, as the price didn't fluctuate too much between earlier datapoints. Even with that said, there definitely seems to be some sort of correlation.

```
sns.residplot(x=data["High"], y=residuals)

sns.residplot(x=data["Low"], y=residuals)

sns.residplot(x=data["Open"], y=residuals)
plt.show()
print("Figure 4: A scatterplot showing overlap in distribution of High, Low and Open variables in the coin_Bitcoin dataset")
```

The above scatterplot shows how similar the scatterplots for the variables High, Low and Open are when compared against residuals. Below these scatterplots are also plotted individually, but to demonstrate the claim that they are similar they have also been plotted together in a single plot.

```
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(8,8))

sns.residplot(ax=ax1, x=data["High"], y=residuals)
ax1.set_title("High vs residuals")

sns.residplot(ax=ax2, x=data["Low"], y=residuals)
ax2.set_title("Low vs residuals")

sns.residplot(ax=ax3, x=data["Open"], y=residuals)
ax3.set_title("Open vs residuals")
plt.tight_layout()
plt.show()
print("Figure 5: Scatterplots of explanatory variables against residuals.")
print("Top: Variable High against residuals. Middle: Variable Low against residuals. Bottom: Variable Open against residuals.")
```

From the above three scatterplots, it is noticable that the distribution of datapoints is very similar between the plots. It is important to point out that the condition of constant variance, which is used to determine the validity of the model, seems to be violated. It is very apparant that once the opening-price surpasses a value of \$20.000, the variance increases a lot as well. This is apparant when we eyeball the average variance up to an opening-price of \$20.000, which seems to be roughly +/- 500, after the opening-price of \$20.000, the variance is much more spread out. Because of this violated condition, we must be more careful when assessing the validity of our final model.

It seems like it would be possible to draw a linear line through the datapoints that is 

```
plt.figure(figsize=(10,10))
sns.scatterplot(range(len(data)), residuals, data=data)
plt.title("Scatterplot of residuals against order of collection")
plt.xlabel("Order of collection")
plt.ylabel("Residuals")
plt.show()
print("Figure 6: A scatterplot of the residuals against the order of collection")
```

What we see in the scatterplot of residuals against the order of collection, it is apparant that the vast majority of the observations are located near the observation made on the day before, and the observation made the day afterwards.It is important to note that the histogram in figure 2 also showed that a very large part of observations has a residual of roughly 0, so in that sense this plot is not surprising. It seems that after roughly 1200 observations the datapoints seem to be more spread out, and after 2500 observations this spread is even bigger. Due to the high volume of datapoints clustered together it is hard to draw a definitive conclusion about the independence of our model. Even though it is not clear cut whether the datapoints are independent of the order of collection or not, our belief is that they are not. If the observations would be truly independent, we would have seen a far bigger spread of datapoints throughout the plot. Because of this, it is adviced to be cautious when attempting to use the final model.

```
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(8,8))

sns.scatterplot(ax=ax1, x=data["High"], y=data["Close"])
ax1.set_title("High vs residuals")

sns.scatterplot(ax=ax2, x=data["Low"], y=data["Close"])
ax2.set_title("Low vs residuals")

sns.scatterplot(ax=ax3, x=data["Open"], y=data["Close"])
ax3.set_title("Open vs residuals")
plt.tight_layout()
plt.show()
print("Figure 7: Scatterplots of explanatory variables against response variable.")
```

Based on the above plots, we see that there is a very clear linear trend between all three predictor variables and the response variable. Based on this, we believe that the assumption of linearity has been met.

### Results

Because we believe that the conditions of our model are reasonably met (with the notion of the beforementioned problems with the plots), we can report the results of our model in the form of a linear equation describing the relation between closing price and our explanatory variables. Our final model will be of the form: $closing price = \beta_0 + \beta_1 \cdot High + \beta_2 \cdot Low + \beta_3 \cdot Open $.

 For our $\beta_0$, which represents the intercept, has a value of $-1.2283$. In the given context, in which the intercept represents the price of Bitcoin if all variables are 0, this is not realistic and should be considered as a theoretical value only. For $\beta_1$ our model has found a coefficient of $0.8874$, which means that for every Dollar increase in the High-price on a given day, the Closing-price will be \$0.8874 higher. For our $\beta_2$ the model has found a coefficient of $0.5532$, which means that for every Dollar higher the Low-price is, the Closing-price will be \$0.5532 higher. Lastly, our model found a coefficient of $-0.4463$ for $\beta_3$. This means that for every Dollar higher Open-price is, that the Closing-price will be \$0.4463 lower.

With the given coefficients, we can now construct the linear equation used to describe the closing price: $closing price = -1.2283 + 0.8874 \cdot High + 0.5532\cdot Low - 0.4463 \cdot Open $ 


### Conclusion and Limitations

Based on our results as described above, we can give an answer to our research question: on a given day, what price indicators can be used to predict the closing-price of Bitcoin? The model suggests that the high-price, low-price and open-price are indicative of the closing price of the same day. Based on the found coefficients, we conclude that the high-price is most influential on the closing price. Contrary to what we initially thought, the variable "volume" does not provide additional explanation about the closing price that was not already provided by the other variables. 
Furthermore, we have found that the "marketcap" variable also does not provide additional explanation. This is surprising, because generally marketcap is based on the price of a product. By excluding the two beforementioned variables, we have not found a noteworthy change in adjusted $R^2$. 

Important to note is that the model has been created based on data that has been collected at the same moment in time. This means that the variables we have collected are published at the same time as the closing-price is. We believe that the model would also be able to give a rough estimate of what the closing-price will be based on the current high- and low-prices on a given day, however this has not been investigated. Furthermore, multiple assumptions of a linear model seem to be violated. Because of this, it is very important to realise that there is likely quite some bias in our model that influences the linear equation used to predict the closing-price.

Based on the model, we have an idea about which variables influence the closing-price of Bitcoin. With this knowledge, we can attempt to 
create similar models for other cryptocurrencies. It is important to note that we must not assume this model is a perfect fit for any cryptocurrency, it should merely serve as a starting point when designing models that describe the closing-price of other currencies.

As a final reminder, the price of Bitcoin is likely not solely dependent on the variables we have mentioned here. As with any other product, market sentiment and a host of other variables are likely to influence change in prices as well. Concretely this means that statements of influential figures can move prices, but this is not captured in our data. Thus our model is based purely on empirical measurements and excludes variables that can not be expressed numerically.
