```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/Tran-Thu-Le/share/main/Time_Series/coins/coin_Bitcoin.csv')
```

```python
price = data["Close"].values
print(price.shape, type(price))
plt.plot(price)
plt.yscale("log")
```

```python
nb_weeks = 155
week_ids = np.arange(nb_weeks)*7
week_price = price[week_ids]1000//7
```

```python
def get_mean(price):
    n = len(price)
    mean_list = []
    mean = price[0]
    for k in range(n-1):
        mean = (k+1)/(k/mean + 1/price[k+1])
        mean_list += [mean]

    mean_np = np.array(mean_list) 
    return mean_np 

my_price = price[1650:]
nb_weeks = len(my_price)//7
id_weeks = np.arange(nb_weeks)*7 
my_price = my_price[id_weeks]

mean = get_mean(my_price) 
plt.plot(my_price, label="price")
plt.plot(mean, label="mean")
plt.legend()
# plt.yscale("log")
```
