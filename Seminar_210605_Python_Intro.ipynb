## Maths

Plot $f(x)$

```python
# Compute f(x)
import numpy as np
x = np.array([-3, -2, -1, 0, 1, 2, 3])
f = x**2
print(f)

# plot f(x)
import matplotlib.pyplot as plt 
plt.plot(x, f)
plt.show()
```

Find derivative $f'(x)$

```python
# Derivative
# Define x
import torch
from torch.autograd import grad
x = torch.linspace(-1., 1., 5, requires_grad = True)
print("x =", x)


# Find derivative
f = x**2
grad = grad(f.sum(), x)[0]
print("derivative = ", grad)
```

# get colors
pc_red = pc[:, :, 0]
pc_green = pc[:, :, 1]
pc_blue = pc[:, :, 2]
plt.imshow(pc_red, cmap="gray")
plt.show()

# change color
pc3 = pc.copy()
pc3[:, :, 0] = pc_blue
pc3[:, :, 1] = pc_green
pc3[:, :, 2] = pc_blue + 50
plt.imshow(pc3)
plt.show()

# split color
m, n, d = pc.shape
line = m//2
pc4 = pc.copy()
pc4[line:, : , :] = pc3[line:, : , :]
plt.imshow(pc4)
